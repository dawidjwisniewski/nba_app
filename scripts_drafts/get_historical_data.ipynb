```python
import json
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
gamenight_date_string = "2010-03-20"
gamenight_date = datetime.strptime(gamenight_date_string, "%Y-%m-%d")
gamenight_date
```

```python
date_for_request = datetime.strftime(gamenight_date, "%Y-%m-%d")
request = requests.get(f"https://www.nba.com/games?date={date_for_request}")
request
```

```python
soup = BeautifulSoup(request.text, 'html.parser')
gamenight_data_json = str(soup.find(id="__NEXT_DATA__").next) # the next element after element with id="__NEXT_DATA__" is the json data 
gamenight_data_dict = json.loads(gamenight_data_json)
```

```python
games_stats_list  = gamenight_data_dict['props']['pageProps']['games']
games_stats_list
```

```python
for i, game in enumerate(games_stats_list):
    print(f"Game {i+1}: {game['homeTeam']['teamName']} vs {game['awayTeam']['teamName']}")
```

```python
for value in games_stats_list[0].values():
    print(type(value))
```

```python
def recursive_keys(multilevel_dictionary: dict, higher_level_key: list[str] = []) -> None:
    """
    Print out all keys in recursive dictionary
    """
    
    key_so_far = higher_level_key
    for key in multilevel_dictionary.keys():
        key_to_pass = key_so_far + [key]
        print_key = True
        
        if isinstance(multilevel_dictionary[key], dict):
            print_key = False
            recursive_keys(multilevel_dictionary[key], key_to_pass)
        if (isinstance(multilevel_dictionary[key], list)) and (len(multilevel_dictionary[key])>0):            
            print_key = False
            recursive_keys(multilevel_dictionary[key][0], key_to_pass)
        else:
            if print_key:
                if higher_level_key == []:
                    print(key)
                else:
                    print("_".join(key_to_pass))
```
