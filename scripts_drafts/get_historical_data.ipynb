```python
def recursive_dict_keys(multilevel_dictionary: dict, higher_level_key: list[str] = []) -> list[str]:
    """
    Print out all keys in recursive dictionary and return list of nested key lists
    
    >>> recursive_keys({'a':1, 'b': {'c':2,'d': 3}})
    [['a'], ['b', 'c'], ['b', 'd']]
    """
    
    keys_list = []
    key_so_far = higher_level_key
    for key in multilevel_dictionary.keys():
        key_to_pass = key_so_far + [key]
        
        if isinstance(multilevel_dictionary[key], dict):
            keys_list.extend(recursive_dict_keys(multilevel_dictionary[key], key_to_pass))
        elif (isinstance(multilevel_dictionary[key], list)) and (len(multilevel_dictionary[key])>0):            
            for i in range(len(multilevel_dictionary[key])):
                keys_list.extend(recursive_dict_keys(multilevel_dictionary[key][0], key_to_pass + [str(i)]))
        else:
            keys_list.append(key_to_pass)
            if higher_level_key == []:
                print(key)
            else:
                print("_".join(key_to_pass))

    return keys_list
```

```python
def convert_types(object_to_check):
    """
    recode Python types into PostgreSQL types strings
    """
    if type(object_to_check) == str:
        return 'VARCHAR'
    elif type(object_to_check) == int:
        return 'integer'
    elif type(object_to_check) == bool:
        return 'boolean'
    elif type(object_to_check) == float:
        return 'numeric'
    elif (object_to_check == []) or (object_to_check is None) :
        return 'None - remove!!!!'
    else:
        return f'{type(object_to_check)} check_manually!!!!'
```

```python
def recursive_dict_keys_and_types(multilevel_dictionary: dict, higher_level_keys: list[str] = []) -> (list[str], list[str]):
    """
    Create two lists:
    - keys in recursive dictionary as string when consecutive levels are denoted with _
    - types of data in recursive dictionary 
    While doing this remove entries with empty data structure (None, [])
    
    """
    keys_list = []
    types_list = []

    keys_so_far = higher_level_keys
    for key in multilevel_dictionary.keys():
        key_to_pass = keys_so_far + [key]
        
        if isinstance(multilevel_dictionary[key], dict):
            #print(recursive_keys(multilevel_dictionary[key], key_to_pass))

            keys_this_level, types_this_level = recursive_dict_keys_and_types(multilevel_dictionary[key], key_to_pass)
            keys_list += keys_this_level
            types_list += types_this_level
        elif (isinstance(multilevel_dictionary[key], list)) and (len(multilevel_dictionary[key])>0):            
            for i in range(len(multilevel_dictionary[key])):
                keys_this_level, types_this_level = recursive_dict_keys_and_types(multilevel_dictionary[key][0], key_to_pass + [str(i)])
                keys_list += keys_this_level
                types_list += types_this_level                
        else:
           # print(type(multilevel_dictionary[key]))
            type_string = convert_types(multilevel_dictionary[key])
            if type_string != 'None - remove!!!!':
                types_list += [type_string]
                #print(types_list)
                if higher_level_keys == []:
                    keys_list += [key]
                else:
                    keys_list += ["_".join(key_to_pass)]
                    #print(keys_list)
               # print(keys_list)

    return (keys_list, types_list)
```

```python
def flatten_recursive_dict(multilevel_dictionary: dict, higher_level_key: list[str] = [], higher_level_dict: dict = {}) -> dict:
    """
    Based on dict with many nested dicts, create dict with one level
    
    # >>> flatten_recursive_dict({'a':1, 'b': {'c':2,'d': 3}, 'e':4})
    # {'a':1, 'b_c':2, 'b_d': 3, 'e': 4}
    """
    key_so_far = higher_level_key
    flat_dict = higher_level_dict

    for key in multilevel_dictionary.keys():
        key_to_pass = key_so_far + [key]
        
        if isinstance(multilevel_dictionary[key], dict):
            flat_dict = flatten_recursive_dict(multilevel_dictionary[key], key_to_pass, flat_dict)
        elif (isinstance(multilevel_dictionary[key], list)) and (len(multilevel_dictionary[key])>0):            
            for i in range(len(multilevel_dictionary[key])):
                flat_dict = flatten_recursive_dict(multilevel_dictionary[key][0], key_to_pass + [str(i)], flat_dict)
        else:           
            if higher_level_key == []:
                flat_dict[key] = multilevel_dictionary[key]
            else:
                flat_dict["_".join(key_to_pass)] = multilevel_dictionary[key]

    return flat_dict
```
